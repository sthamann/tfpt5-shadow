# 03 — Template for real EHT data

Walks the same pipeline through a real polarimetric image cube. This notebook is a *template*: replace the data-loading cells with paths to your local copy of the EHT 2018 polarimetric release.

## Required input shape

* `x, y`: 2-D coordinate grids in units of the gravitational radius $r_g$.
* `lambda_sq`: 1-D array of $\lambda^2$ values per frequency channel (m\u00b2).
* `chi`: 3-D polarization-angle cube of shape `(F, H, W)` in radians.
* `sigma_chi`: per-pixel angle uncertainty, same shape.
* `chi_grmhd`: 3-D GRMHD forward-model cube, same shape (or supply only `chi_0_grmhd` if you trust an external RM fit).

## Steps

1. Load `PolarimetricImage` from EHT release.
2. Load GRMHD model from your `ipole`/`harmpi` run.
3. Fit per-pixel RM with `rotation_measure_fit`.
4. Compute the residual intercept.
5. Deproject radially around the photon-ring barycentre.
6. Repeat with reversed $E\!\cdot\!B$ orientation for the third null.
7. Call `run_all_nulls` and write the report to `results/`.
